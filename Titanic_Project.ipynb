### Question 

- What is the total number of passengers in each passenger class (Pclass)?

- How many passengers are children (e.g., Age < 18) and how many are adults?

- What is the count of passengers who embarked from each port (Embarked)?

- What are the top 5 highest fares paid, and who paid them (Name, Pclass, and Cabin)?

- How many passengers were assigned to each deck (based on the first letter of Cabin)?

```python
# What is the total number of passengers in each passenger class (Pclass)?
P_count = data['Pclass'].value_counts()  

plt.bar(P_count.index.astype(str), P_count.values, color='skyblue')  
plt.xlabel('Passenger Class')  
plt.ylabel('Count of Passengers')  
plt.title('Histogram of Passenger Count by Class')  
plt.show() 
```

```python
# How many passengers are children (e.g., Age < 18) and how many are adults?
Child_count = len(data[data['Age'] <= 18])
print(f"Number of chidrens who are 18 or less in Age: {Child_count}")
```

```python
# What is the count of passengers who embarked from each port (Embarked)?
# Southampton (S)
# Cherbourg (C)
# Queenstown (Q)
embarked_count = data['Embarked'].value_counts()
embarked_count
```

```python
Max_fare = data[data['Fare'] == data['Fare'].max()]
result = Max_fare[['Name', 'Pclass', 'Cabin', 'Fare']].head(5)
result
```

```python
#How many passengers were assigned to each deck (based on the first letter of Cabin)?
data['Cabin'] = data['Cabin'].dropna().str[0]

#Lineplot
deck_count = data['Cabin'].value_counts()
deck_count.sort_index().plot(kind='line', marker='o')  
plt.title('Number of Passengers Assigned to Each Deck')  
plt.xlabel('Deck')  
plt.ylabel('Number of Passengers')  
plt.grid()  
plt.show()  
```
